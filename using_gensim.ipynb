```python
import gensim
from gensim.utils import simple_preprocess
import pandas as pd
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')

import gensim.downloader as api

wv = api.load('word2vec-google-news-300')
```

```python
df=pd.read_csv("IMDB_Dataset.csv")
```

```python
df.head()
```

```python
df["sentiment"]=pd.get_dummies(df["sentiment"],drop_first=True)

df["sentiment"].head()
```

```python
type(len(df))
```

```python
from nltk.stem.porter import PorterStemmer
import re
from nltk.stem import WordNetLemmatizer
from tqdm import tqdm
from nltk.corpus import stopwords
corpus=[]

lem=WordNetLemmatizer()
ste=PorterStemmer()
#df=nltk.sent_tokenize(df,language='english')

for i in tqdm(range(0,len(df))):
  sent=re.sub('[^a-zA-Z]',' ',df['review'][i])
  sent=sent.lower()
  sent=sent.split()
  sent=[lem.lemmatize(word) for word in sent if not word in set(stopwords.words('english')) ]
  sent=' '.join(sent)
  corpus.append(sent)

```

```python
corpus[0]
```

```python
from nltk.corpus.reader.tagged import sent_tokenize
words=[]

for sent in corpus:
  sent_token=sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))

```

```python
words
```

```python

model=gensim.models.Word2Vec(words,window=5,min_count=2)
```

```python
model.wv.similar_by_word('end')
```

```python
import numpy as np

def Avg_word2vec(dec):

  return np.mean([model.wv['word'] for word in dec if word in model.wv.index_to_key],axis=0)
```

```python
from tqdm import tqdm

x=[]

for i in tqdm(range(len(words))):
  x.append(Avg_word2vec(words[i]))
```

```python
X_new=np.array(x)
```

```python
X_new
```

```python
y=df["sentiment"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_new,y,test_size=0.33,random_state=42)
```

```python
print("shape of x_train" ,X_train.shape)
print("shape of X_test" ,X_test.shape)
print("shape of y_train" ,y_train.shape)
print("shape of y_test" ,y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,accuracy_score
```

```python
model=RandomForestClassifier()

model.fit(X_train,y_train)
pred_y=model.predict(X_test)
print(classification_report(y_test,pred_y))
print("Accuracy score is",accuracy_score(y_test,pred_y))
```

```python
from sklearn.tree import DecisionTreeClassifier

model_m=DecisionTreeClassifier()

model_m.fit(X_train,y_train)
pred_y=model_m.predict(X_test)



print(classification_report(y_test,pred_y))

print("Accuracy score is",accuracy_score(y_test,pred_y))
```

```python

from sklearn.linear_model import LogisticRegression

model_m=LogisticRegression()

model_m.fit(X_train,y_train)
pred_y=model_m.predict(X_test)



print(classification_report(y_test,pred_y))

print("Accuracy score is",accuracy_score(y_test,pred_y))
```
